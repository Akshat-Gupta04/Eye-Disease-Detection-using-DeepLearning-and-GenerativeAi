```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix
import os

# Load the dataset
train_dir = '/Users/akshat/Documents/Model Design AI/fundus_final_completed/new_output/train'
val_dir = '/Users/akshat/Documents/Model Design AI/fundus_final_completed/new_output/val'
test_dir = '/Users/akshat/Documents/Model Design AI/fundus_final_completed/new_output/test'

# Data Augmentation and Preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Load datasets
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=1,
    class_mode='categorical',
    shuffle=False
)

# List of Pre-trained models to be used
pretrained_models = {
    "InceptionV3": tf.keras.applications.InceptionV3,
    "ResNet50": tf.keras.applications.ResNet50,
    "EfficientNetB0": tf.keras.applications.EfficientNetB0,
    "EfficientNetB4": tf.keras.applications.EfficientNetB4,
    "DenseNet121": tf.keras.applications.DenseNet121,
    "VGG16": tf.keras.applications.VGG16,
    "MobileNetV2": tf.keras.applications.MobileNetV2
}

# Create a custom model function
def create_model(base_model_class, input_shape=(224, 224, 3), num_classes=10):
    # Load the pre-trained base model without the top (classification) layer
    base_model = base_model_class(
        weights='imagenet',  # Load ImageNet pre-trained weights
        include_top=False,   # Don't include the final classification layer
        input_shape=input_shape
    )
    
    # Freeze the base model layers so they aren't updated during training
    base_model.trainable = False

    # Add custom layers for the new classification task
    model = tf.keras.Sequential([
        base_model,
        tf.keras.layers.GlobalAveragePooling2D(),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(num_classes, activation='softmax')  # Adjust num_classes to your dataset
    ])

    # Compile the model
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# Training parameters
epochs = 30
batch_size = 32
models_trained = {}
histories = {}

# Loop over each pre-trained model and train it
for model_name, model_class in pretrained_models.items():
    print(f"\nTraining model: {model_name}")

    # Create a model using the current pre-trained model
    model = create_model(model_class)

    # Define early stopping and learning rate reduction callbacks
    early_stopping = tf.keras.callbacks.EarlyStopping(
        monitor='val_loss', patience=5, restore_best_weights=True, verbose=1
    )
    
    reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', factor=0.1, patience=3, min_lr=1e-6, verbose=1
    )

    # Train the model
    history = model.fit(
        train_generator,
        epochs=epochs,
        validation_data=val_generator,
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )

    # Store the trained model and training history
    models_trained[model_name] = model
    histories[model_name] = history

    # Evaluate the model on the test set
    print(f"\nEvaluating {model_name} on test set:")
    test_loss, test_accuracy = model.evaluate(test_generator)
    print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

    # Generate confusion matrix and classification report
    y_true = test_generator.classes
    y_pred = np.argmax(model.predict(test_generator), axis=1)

    print(f"Confusion Matrix for {model_name}:")
    print(confusion_matrix(y_true, y_pred))

    print(f"Classification Report for {model_name}:")
    print(classification_report(y_true, y_pred, target_names=test_generator.class_indices.keys()))

# Plot accuracy and loss graphs for each model
for model_name, history in histories.items():
    # Accuracy Plot
    plt.figure(figsize=(10, 5))
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_name} Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

    # Loss Plot
    plt.figure(figsize=(10, 5))
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```

```python
import pandas as pd

# Dictionary to store metrics for the table
model_metrics = {
    "Model Name": [],
    "Train Accuracy": [],
    "Validation Accuracy": [],
    "Train Loss": [],
    "Validation Loss": [],
    "Test Accuracy": [],
    "Test Loss": []
}

# Loop over trained models to collect metrics
for model_name, model in models_trained.items():
    # Extract metrics from the history object
    history = histories[model_name]

    # Get the last epoch's training and validation accuracy/loss
    train_accuracy = history.history['accuracy'][-1]
    val_accuracy = history.history['val_accuracy'][-1]
    train_loss = history.history['loss'][-1]
    val_loss = history.history['val_loss'][-1]

    # Evaluate the model on the test set to get test accuracy and loss
    test_loss, test_accuracy = model.evaluate(test_generator)

    # Store the metrics for this model
    model_metrics["Model Name"].append(model_name)
    model_metrics["Train Accuracy"].append(train_accuracy)
    model_metrics["Validation Accuracy"].append(val_accuracy)
    model_metrics["Train Loss"].append(train_loss)
    model_metrics["Validation Loss"].append(val_loss)
    model_metrics["Test Accuracy"].append(test_accuracy)
    model_metrics["Test Loss"].append(test_loss)

# Create a pandas DataFrame to display the summary table
df_metrics = pd.DataFrame(model_metrics)

# Display the summary table
print("\nSummary of Model Performance:")
print(df_metrics)

# Optionally, you can save the table as a CSV file for later reference
df_metrics.to_csv('model_performance_summary.csv', index=False)
```
